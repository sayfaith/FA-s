```python
#5

df['Likes'] = df['Likes'].fillna(0)
df['Comments'] = df['Comments'].fillna(0)
df.dropna(inplace=True)
```

```python
#6

df.describe()
```

```python
#7

album_types = df['Album_type'].value_counts()
album_types
```

```python
#8

plt.pie(album_types, labels=album_types.index, autopct='%1.1f%%')
plt.title('Album Types')
plt.legend(album_types.index, loc='best')

plt.show()
```

<b>From the above pie chart, we can make the following conclusions:</b>

‚Ä¢ The majority of the songs in the dataset are albums, accounting for more than 70% of the total songs.

‚Ä¢ The second most common type of album is a single, accounting for almost 30% of the total songs, followed by compilation, which is the least common type of album with only 3.6% of the total songs.

‚Ä¢ This suggests that albums are the most popular type of release among the artists in the dataset, and that artists in the dataset tend to release albums more frequently than they do singles or compilations.

```python
#14

enerdance = df[['Track', 'Energy', 'Danceability']]
enerdance = enerdance.sort_values(by=['Energy', 'Danceability'], ascending=False)

enerdance
```

```python
#15

music_attributes = df.sort_values('Stream', ascending=False).head(10)

attributes = ['Energy', 'Danceability', 'Acousticness']
data = music_attributes[['Track'] + attributes].set_index('Track')

ax = data.plot(kind='bar', figsize=(10, 6), width=0.8)

ax.set_title('Top 10 Tracks by Stream Count and Attributes')
ax.set_xlabel('Track')
ax.set_ylabel('Value')
ax.legend(attributes)

plt.show()

```

<b> From the above bar chart, we can make the following conclusions: </b>

‚Ä¢ Danceability are generally taller than the bars for the other attributes, suggesting that the top 10 tracks are generally more danceable than they are energetic or acoustic

‚Ä¢ For higher chances of charting in music streams, artists should consider releasing songs that are danceable and energetic.

```python
#16 

topten = df.groupby('Artist')['Likes'].sum().nlargest(10)
plt.figure(figsize=(12,6))
plt.bar(topten.index, topten.values)
plt.title('Number of Likes per Artist')
plt.xlabel('Likes')
plt.ylabel('Artist')
plt.xticks(rotation=45)

plt.show()
```

<b> From the above bar chart, we can make the following conclusions: </b>

‚Ä¢ BTS and Blackpink are the most favored artists on Spotify, as indicated by the number of likes they have garnered.

‚Ä¢ The rest of the other artists are far behind and closely ranked.

```python
#17

singles = df[df['Album_type'] == 'single'].groupby('Artist').agg({'Views': 'sum', 'Likes': 'sum'}).sort_values(by='Likes', ascending=False).head(10)

singles
```

```python
#19

compilations = df[df['Album_type'] == 'compilation'].groupby('Artist').agg({'Views': 'sum', 'Likes': 'sum'}).sort_values(by='Likes', ascending=False).head(10)

compilations
```

```python
#20

top_channels = df.groupby('Channel')['Views'].sum().nlargest(10)
plt.figure(figsize=(12,6))
plt.bar(top_channels.index, top_channels.values)
plt.title('Top Spotify Channels by Views')
plt.xlabel('Channel')
plt.ylabel('Total Views')
plt.xticks(rotation=45)

plt.show()
```

<b> From the above bar chart, we can make the following conclusions: </b>

‚Ä¢ T-Series, which has accumulated 50 billion views is the most widely watched channel.

‚Ä¢ LuisFonsiVEVO, CalvinHarrisVEVO, Macklemore and the remaining channels within the top 10 are far behind with T-series but are closely ranked with each other.
